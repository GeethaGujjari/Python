#**Dictionaries**

```python
L1= [1,2,2,2,3,3,4,4,4,4,5,6,6,8,9,10]
dict={}
for i in L1:
  if i in dict:
    dict[i]+=1
  else:
    dict[i]=1

print(dict)
temp=0
for k,v in dict.items():
  if v>temp:
    temp=v
    ans=k
print("The key is ",ans)
```

```python
L1= [1,2,2,2,3,3,4,4,4,4,5,6,6,8,9,10]
max_count=0
for i in L1:
  count=L1.count(i)
  if count>max_count:
    max_count=count
    ans=i
print("i =",ans," count = ",max_count)
```
